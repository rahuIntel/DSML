```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
plt.plot(x,y)
plt.grid()
plt.xlabel("X Label for the Plot")
plt.ylabel("y Label for the Plot")
plt.title("Title for the Plot")
```

```python
x = np.arange(0.0,25.0,0.1)
plt.plot(x,np.cos(x))
plt.grid()
plt.xlabel("X Label for the Plot")
plt.ylabel("y Label for the Plot")
plt.title("Title for the Plot")
```

```python
plt.figure(figsize=(15,4))
x = np.arange(0.0,25.0,0.1)
plt.subplot(1,3,1)
plt.plot(x,np.cos(x))
plt.grid()
plt.xlabel("X Label for the Cos Plot")
plt.ylabel("y Label for the Cos Plot")
plt.title("Title for the Cos Plot")
plt.subplot(1,3,2)
plt.plot(x,np.sin(x))
plt.grid()
plt.xlabel("X Label for the Sin Plot")
plt.ylabel("y Label for the Sin Plot")
plt.title("Title for the Sin Plot")
plt.subplot(1,3,3)
plt.plot(x,np.tan(x))
plt.grid()
plt.xlabel("X Label for the Sin Plot")
plt.ylabel("y Label for the Sin Plot")
plt.title("Title for the Tan Plot")
```

```python
plt.figure(figsize=(15,7))
plt.subplot(2,2,1)
x = [1,2,3,4,5,6]
y = [22,33,44,33,25,66]
tick_labels = ['o','t','th','f','ff','si']
plt.bar(x,y,tick_label=tick_labels, width = 0.2)
plt.xlabel("X Label for the Sin Plot")
plt.ylabel("y Label for the Sin Plot")
plt.title("Title for the BAR Plot")

plt.subplot(2,2,2)
x = [1,2,3,4,5,6]
y = [22,33,44,33,25,66]
tick_labels = ['o','t','th','f','ff','si']
plt.bar(x,y,tick_label=tick_labels, width = 0.5, color='r')
plt.xlabel("X Label for the Sin Plot")
plt.ylabel("y Label for the Sin Plot")
plt.title("Title for the BAR Plot")

plt.subplot(2,2,3)
x = [1,2,3,4,5,6]
y = [22,33,44,33,25,66]
tick_labels = ['o','t','th','f','ff','si']
plt.bar(x,y,tick_label=tick_labels, width = 1.0, color='g')
plt.xlabel("X Label for the Sin Plot")
plt.ylabel("y Label for the Sin Plot")
plt.title("Title for the BAR Plot")

plt.subplot(2,2,4)
x = [1,2,3,4,5,6]
y = [22,33,44,33,25,66]
tick_labels = ['o','t','th','f','ff','si']
plt.bar(x,y,tick_label=tick_labels, width = 0.8, color=['g','r','b'])
plt.xlabel("X Label for the Sin Plot")
plt.ylabel("y Label for the Sin Plot")
plt.title("Title for the BAR Plot")
```

```python
# Histogram
```

```python
plt.figure(figsize=(10,10))

plt.subplot(3,1,1)
age = [2,34,23,56,55,43,56,67,77,66,45,46,78,99,66,44,54,25,47,3,5,78,44,32,45,67,54,43,56,33,56,43,67,55,33,67,9,21,34]
range = (0,100)
bin = 5

plt.hist(age,bin, range, histtype='bar', rwidth=0.7, align="right",color='y', stacked=True)
plt.xlabel("Age")
plt.ylabel("Y Axis")
plt.title("Ages")

plt.subplot(3,1,2)
counts, bins = np.histogram(age)
plt.hist(bins[:-1], bins, weights=counts)

plt.subplot(3,1,3)
plt.bar(age,height=np.arange(0,len(age)))
```

```python
Temp = [12,23,35,56,77,78]
sale = [3,6,7,8,9,11]
plt.grid()
plt.scatter(Temp, sale, label="Sales", color="green", marker = "*", s=100)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.legend()
plt.title('Scatter Graph')
```

```python
# Pie Plot

activities = ['Eat','Sleep','Work', 'Play']
slices = [3,5,13,3]
colors = ['b','g','r','y']

plt.pie(slices, labels=activities,colors=colors, startangle=90, shadow=True, explode=(0,0,0,0.2), autopct='%1.2f%%')
plt.legend()
```
